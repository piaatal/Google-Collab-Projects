I created this model with social media apps like instagram in mind. Apps like instagram thrive on thousands of visually captivating pictures being posted everyday. Instagram offers the ability for users to add hidden captions to their pictures, and essentially the purpose of this is to describe exactly what's happening in the picture.

These descriptions serve two main purposes: inclusivity and marketing. As for inclusivity, if people are visually impaired or unable to view the picture that is posted, then instead the can view or listen to the user give description. Secondly, it is helpful for marketing purposes because instagram uses key words from the description to help the posts show up on people's feeds.

The descriptions are written by the user and it is often time consuming to write descriptions for each and every single picture in sometimes twenty picture posts. That's where this AI model comes in. Using the pre-trained Saleforce image to text model on hugging face, I used the pipeline function to input two pictures that I wanted to post. The model did a very good job of accurately coming up with a caption for these two images and I can now use them when I post on social media. I just inputted two pictures, but I can input as many pictures as I want in my drive and input them into the model for the model to loop through and come up with captions for.

```python
from transformers import pipeline

model_name = "Salesforce/blip-image-captioning-large"

classifier = pipeline(task="image-to-text", model=model_name)

classifier
```

```python
pictures = ['IMG_4818.jpg','IMG_4786.jpg']
```

```python
from PIL import Image

for image in pictures:
  image_path = '/content/drive/My Drive/Picture'
  image_path = os.path.join(image_path, image)
  image = Image.open(image_path).convert("RGB")
  caption = classifier(image)
  print("Generated Caption:", caption[0]['generated_text'])
```
